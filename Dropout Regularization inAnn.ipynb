```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("/content/gdrive/MyDrive/sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.shape
```

```python
#df.isnull().sum()
df.isna().sum()
```

```python
#print column names
df.columns
```

```python
#how many values we have in column '60'
df[60].value_counts()
```

```python
X = df.drop(60, axis=1)
y = df[60]
y.head()
```

```python
X
```

```python
#we need to do onehotencoding for column '60' because it has categorical variable
y = pd.get_dummies(y, drop_first=True)
```

```python
y.sample(5) # R -->1 and M-->0
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1)
```

```python
#Check the shape
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
        keras.layers.Dense(60, input_dim =60, activation='relu'),
        keras.layers.Dense(30, activation = 'relu'),
        keras.layers.Dense(15, activation = 'relu'),
        keras.layers.Dense(1, activation= 'sigmoid')
])

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred[:5] # it is in 2 dimensional shape
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer i.e 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
modeld = keras.Sequential([
        keras.layers.Dense(60, input_dim =60, activation='relu'),
        keras.layers.Dropout(0.5), #here we drop 50%
        keras.layers.Dense(30, activation = 'relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(15, activation = 'relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(1, activation= 'sigmoid')
])

modeld.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```
